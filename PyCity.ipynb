# PyCity Schools Analysis

* As a whole, schools with higher budgets, did not yield better test results. By contrast, schools with higher spending per student actually (\$645-675) underperformed compared to schools with smaller budgets (<\$585 per student).

* As a whole, charter schools out-performed the public district schools across all metrics. However, more analysis will be required to glean if the effect is due to school practices or the fact that charter schools tend to serve smaller student populations per school.

```python
#Calculate the total number of schools
#school_data_total = len(school_data)

school_data_total = len(school_data)
print(school_data_total)
```

```python
#Calculate the total number of students
student_data_total = len(student_data)

print(student_data_total)
```

```python
#Calculate the total budget
school_budget = school_data["budget"].sum()
school_budget_fmt = ("$" + format(school_budget, ',.2f'))
print(school_budget_fmt)
```

```python
#Calculate the average math score
avg_math_score = round(student_data["math_score"].mean())
avg_math_score_fmt = format(avg_math_score, ',.2f')
print(avg_math_score_fmt)
```

```python
#Calculate the average reading score
avg_reading_score = round(student_data["reading_score"].mean())
avg_reading_score_fmt = format(avg_reading_score, ',.2f')
print(avg_reading_score_fmt)
```

```python
#Calculate the overall passing rate (overall average score), i.e. (avg. math score + avg. reading score)/2

overall_pass_rate = round(avg_reading_score + avg_math_score)/2
overall_pass_rate_fmt = format(overall_pass_rate, ',.2f')
print(overall_pass_rate_fmt)

overall_pass_rate_pct = str(overall_pass_rate_fmt) + "%"
overall_pass_rate_pct 
```

```python
#Calculate the percentage of students with a passing math score (70 or greater)
math_gt_70_df = student_data.loc[student_data["math_score"] >= 70].count()["student_name"] 
math_pct_gt_70 =  (math_gt_70_df / student_data_total) * 100
math_pct_gt_70_fmt = format(math_pct_gt_70,',.2f')
math_pct_gt_70_fmt = str(math_pct_gt_70_fmt) + "%"
                            
math_pct_gt_70_fmt
```

```python
#Calculate the percentage of students with a passing math score (70 or greater)
read_gt_70_df = student_data.loc[student_data["reading_score"] >= 70].count()["student_name"]
read_pct_gt_70 =  (read_gt_70_df / student_data_total) * 100 
read_pct_gt_70_fmt = format(read_pct_gt_70,',.2f')
read_pct_gt_70_fmt = str(read_pct_gt_70_fmt) + "%"

read_pct_gt_70_fmt
```

```python
summary_df = pd.DataFrame({"Total Schools": [school_data_total],
                           "Total Students":[student_data_total],
                           "Total Budget":[school_budget_fmt],
                           "Average Math Score":[avg_math_score_fmt],
                           "Average Reading Score":[avg_reading_score_fmt],
                           "% Passing Math":[math_pct_gt_70_fmt],
                           "% Passing Reading":[read_pct_gt_70_fmt],
                           "% Overall Passing Rate":[overall_pass_rate_pct]
                          })

summary_df
```

```python
#create index on school_name
#school_list_types = school_data_complete.set_index(["school_name"])["type"]
school_list_typs = school_data.set_index(["school_name"])["type"]
school_list = school_data_complete.groupby(['school_name']) ["type"]

school_list_typs.head(20)
```

```python
#groupby school name and size (count) of school
school_count = school_data_complete.groupby(["school_name"]) ["type"].count()
print(school_count)
```

```python
#average math score by each school
school_math_mean = school_data_complete.groupby(["school_name"]) ["math_score"].mean()
print(school_math_mean)
```

```python
#average reading score by each school
school_read_mean = school_data_complete.groupby(["school_name"]) ["reading_score"].mean()
print(school_read_mean)
```

```python
#budget for each school
school_bud = school_data_complete.groupby(["school_name"]) ["budget"].mean() 
print(school_bud)
```

```python
#cost per student for each school
per_student_budget = (school_bud / school_count) 
print(per_student_budget)
```

```python
#students with math score >= 70 or reading score >= 70
school_pass_math = school_data_complete.loc[school_data_complete["math_score"] >= 70]

school_pass_math_grouped = school_pass_math.groupby(["school_name"]).count() ["student_name"]

school_pass_read = school_data_complete.loc[school_data_complete["reading_score"] >= 70]
school_pass_read_grouped = school_pass_read.groupby(["school_name"]).count() ["student_name"]

print(school_pass_math_grouped, school_pass_read_grouped)

#school_pass_math_grouped
```

```python
#pct passing reading
pct_pass_read = (school_pass_read_grouped / school_count) * 100 
print(pct_pass_read)
```

```python
#pct passing math
pct_pass_math = (school_pass_math_grouped / school_count) * 100 
print(pct_pass_math)
```

```python
#overall passing rate
overall_passing_rt = (pct_pass_read  + pct_pass_math ) / 2
overall_passing_rt
```

```python
school_bud = school_bud.map("${:,.2f}".format)
per_student_budget = per_student_budget.map("${:,.2f}".format)
```

```python
summary_school_df = pd.DataFrame({"School Type": school_list_typs,
                          "Total Students": school_count,
                           "Total Budget": school_bud,
                           "Per Student Budget" : per_student_budget,
                           "Average Math Score": school_math_mean,
                           "Average Reading Score": school_read_mean,
                           "% Passing Math": pct_pass_math,
                           "% Passing Reading": pct_pass_read,
                           "% Overall Passing Rate": overall_passing_rt
                          })
summary_school_df
```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Math Score
  * Average Reading Score
  * % Passing Math

```python
# Sample bins. Feel free to create your own bins.
spending_bins = [0, 585, 615, 645, 675]
group_names = ["<$585", "$586-615", "$616-645", "$646-675"]
```

```python
# Remove percent formatting from "Per Student Budget" columns and convert to float in order to use cut method

summary_school_df["Per Student Budget"] = summary_school_df["Per Student Budget"].str.replace(',', '')
summary_school_df["Per Student Budget"] = summary_school_df["Per Student Budget"].str.replace('$', '')
summary_school_df["Per Student Budget"] = summary_school_df["Per Student Budget"].astype(float)
```

```python
# Duplicate data frame - need to use copy function, otherwise original df is also affected when slicing
# Perform cut and drop unnecessary columns

spending_range_df = summary_school_df.copy()
spending_range_df["Spending Range"] = pd.cut(spending_range_df["Per Student Budget"], spending_bins, labels=group_names)
spending_range_df.drop(['School Type','Total Budget', 'Per Student Budget'], axis=1, inplace=True)

spending_range_df
```

```python
# Group by Spending Range and create variables for dataframe

spending_range_df_grouped = spending_range_df.groupby("Spending Range")

spending_math = spending_range_df_grouped["Average Math Score"].mean()

spending_read = spending_range_df_grouped["Average Reading Score"].mean()

spending_pass_math = spending_range_df_grouped["% Passing Math"].mean()

spending_pass_read = spending_range_df_grouped["% Passing Reading"].mean()

spending_pass_overall = spending_range_df_grouped["% Overall Passing Rate"].mean()


print(spending_pass_read)

#spending_range_df_grouped_sort.head(20)
```
